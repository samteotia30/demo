```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Age": [25,34,45,23,54,32,40,27,50,33],
    "Income (in $1000s)": [40,60,80,30,120,55,75,50,100,45],
    "Total Spending (in $1000s)": [10,15,20,12,25,18,22,16,28,20],
    "Frequency of Visits": [12,20,30,8,50,18,28,15,35,14],
    "Products Purchased": [5,8,12,3,18,7,11,6,16,7]
}

df = pd.DataFrame(data)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

#Ensuring we get enough components
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

pca_df = pd.DataFrame(pca_result, columns=["PC1","PC2"])

plt.figure(figsize=(8,6))
plt.scatter(pca_df["PC1"],pca_df["PC2"],c="blue", edgecolor='k', s=100)
plt.title('PCA of Customer Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```
