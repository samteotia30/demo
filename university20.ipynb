```python
import pandas as pd
df = pd.read_csv("/content/world all university rank and rank score.csv")
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
import statsmodels.api as sm  # Assuming you meant 'statsmodels' instead of 'scripmodel'

from IPython import get_ipython
from IPython.display import display
from scipy.stats import chisquare # Import chisquare for Chi-square test

# Print the available columns in your DataFrame:
print(df.columns)

# Choose the correct column name based on the output above
# and assign it to col_name:
# col_name = 'Rank'  # Example: If 'Rank' is the correct column name
# Instead of assuming 'Rank' is the column name, let's look for a similar name
# assuming your column names are all lowercase or use a mix of upper and lowercase,
# let's make the name lowercase and check
col_name = 'rank' # Example: If 'rank' is the correct column name

# if this fails try searching for similar strings
# for col in df.columns:
#   if 'rank' in col.lower():  # check if rank is in the name in a case-insensitive way
#     col_name = col
#     break
# This block is commented out to avoid execution unless it's the appropriate fix

global_mean = df[col_name].mean()
sample_mean = df[col_name].mean()
population_std = df[col_name].std()
sample_std = df[col_name].std()
sample_size = len(df[col_name])

print(f"Global Mean: {global_mean}")
print(f"Sample Mean: {sample_mean}")
print(f"Population Standard Deviation: {population_std}")
print(f"Sample Standard Deviation: {sample_std}")
print(f"Sample Size: {sample_size}")

# --- Chi-square Test for Goodness of Fit ---

# 1. Define expected frequencies (assuming uniform distribution)
unique_ranks = df[col_name].nunique()
expected_freq = len(df[col_name]) / unique_ranks

# 2. Calculate observed frequencies
observed_freq = df[col_name].value_counts().sort_index().values

# 3. Perform Chi-square test
chi2_statistic, p_value = chisquare(observed_freq, f_exp=expected_freq)

print(f"\nChi-square Statistic: {chi2_statistic}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The distribution of ranks is significantly different from uniform.")
else:
    print("Fail to reject the null hypothesis: The distribution of ranks is not significantly different from uniform.")
```
