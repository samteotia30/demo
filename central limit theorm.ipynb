```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#step 1:generate a population with a non-normal distribution (uniform distribution)
population_size=10000
population=np.random.normal(0,100,size=population_size)
```

```python
#plot the population distribution
plt.figure(figsize=(10,6))
sns.histplot(population,bins=30,kde=True,color='skyblue')
plt.title('Population Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
#step 2: set parameters for the central limit theorm distribution
sample_size=30   #size of each sample
num_samples=1000  #number of samples to draw
```

```python
#step 3:draw multiple samples and calculate their means
sample_means = []
for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=True)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)
```

```python
#step 4:
plt.figure(figsize=(10,6))
sns.histplot(sample_means,bins=30,kde=True,color='lightgreen')
plt.title('Central Limit Theorem Distribution')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```
