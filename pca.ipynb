```python
import pandas as pd

data = {
    "Study Hours": [5, 3, 4, 6, 5, 2, 3, 4, 5, 6],
    "Recreation Hours": [2, 4, 3, 1, 2, 5, 4, 3, 2, 1],
    "Attendance (X)": [90, 75, 80, 95, 85, 60, 76, 75, 85, 95],
    "Assignment Submitted": [8, 5, 6, 9, 7, 3, 4, 6, 7, 9],
    "Sleep Hours": [7, 6, 6, 5, 8, 7, 5, 5, 5, 6] # Corrected the length of 'Sleep Hours' to 10
}

df = pd.DataFrame(data)
print(df)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])
print("Explained Variance Ratio:")
print( pca.explained_variance_ratio_)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.scatter(pca_df["PC1"],pca_df["PC2"],c="blue", edgecolor='k', s=100)
plt.title('PCA of Academic Performance Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```
