```python
#question 3
import pandas as pd
df=pd.read_csv('/content/titanic (2).csv')
df.head()
```

```python
#mean of the fares
mean_fares=df['Fare'].mean()
print(mean_fares)
```

```python
#standard deviation
standard_deviation=df['Fare'].std()
print(standard_deviation)
```

```python
#histogram
import matplotlib.pyplot as plt
plt.hist(df['Fare'],bins=10)
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Histogram of Fare')
plt.show()
```
