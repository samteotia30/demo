```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#step 1:generate a population with a non-normal distribution (uniform distribution)
population_size=10000
population=np.random.uniform(0, 100, population_size)
```

```python
#plot the population distribution
plt.figure(figsize=(12, 5))
sns.histplot(population,bins=30,kde=True,color='skyblue')
plt.title('Population Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
#step 2: set parameters for the central limit theorm distribution
sample_size=30   #size of each sample
num_samples=1000  #number of samples to draw
```

```python
#step 3:draw multiple samples and calculate their means
sample_means = []
for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=True)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)
```

```python
#step 4:
plt.figure(figsize=(10,6))
sns.histplot(sample_means,bins=30,kde=True,color='lightgreen')
plt.title('Central Limit Theorem Distribution')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```

```python
#define the original population distribution
#for thos example,our populaton is a uniform distribution represented
#by the possible outcomes of a
#6-sided dice (1 through 6)

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#possible outcomes for a 6-sided dice
dice_outcomes=np.array([1,2,3,4,5,6])
```

```python
#visualize the uniform distribution of dice rolls
plt.hist(dice_outcomes,bins=6,edgecolor='black',align='left',color='lightgreen',rwidth=0.8)
plt.xticks(dice_outcomes)
plt.title('Uniform Distribution of Dice Rolls')
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.show()
```

```python
#draw random samples and calculate sample means
#set a sample size(e.g., n=10) and roll the dice multiple times to create sample
sample_size=10
num_samples=1000
sample_means=[]
for _ in range(num_samples):
    sample=np.random.choice(dice_outcomes,size=sample_size,replace=True)
    sample_mean=np.mean(sample)
    sample_means.append(sample_mean)
```

```python
#plot the distribution of sample means
#plot the histogram of the sample means
#as you increase the number of samples and samples size,the distribution
#of sample means should start to resemble
# a normal distribution
plt.hist(sample_means,bins=20,edgecolor='black',align='left',color='lightgreen',rwidth=0.8)
plt.title('Distribution of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```
